```python
from tensorflow.keras.preprocessing.text import  one_hot
```

```python
sent=['the glass of milk',
      'the glass of juice',
      'the cup of tea',
      'I am a good boy',
      'I am a good developer',
      'understand the meaning of words',
      'your videos are good']
```

```python
sent
```

```python
voc_size=10000
```

```python
one_hot_rep=[one_hot(sentence,voc_size) for sentence in sent]
one_hot_rep
```

```python
#word embedding representation

from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
#pad_sequences is used to equal length of words in each sentance 
sent_lenght=8 
```

```python
embbed_sent=pad_sequences(one_hot_rep,padding="pre",maxlen=sent_lenght)
embbed_sent
```

```python
#feature representation
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_lenght))
model.compile("adam","mse")
```

```python
model.summary()
```

```python
model.predict(embbed_sent)
```
